```python
import graphlab
```

```python
image_train = graphlab.SFrame('image_train_data')
```

```python
image_train.head()
```

# Training a nearest-neighbor model for image retrieval using deep features

```python
knn_model = graphlab.nearest_neighbors.create(image_train, features=['deep_features'], label='id')
```

# Using image retrieval model with deep features to find similarity

```python
graphlab.canvas.set_target('ipynb')
cat = image_train[18:19]
cat['image'].show()
```

```python
knn_model.query(cat)
```

```python
def get_images(result):
    return image_train.filter_by(result['reference_label'], 'id')
```

```python
cat_neighbors = get_images(knn_model.query(cat))
```

```python
cat_neighbors['image'].show()
```

```python
car = image_train[8:9]
car['image'].show()
```

```python
get_images(knn_model.query(car))['image'].show()
```

```python
bird = image_train[138:139]
bird['image'].show()
```

```python
get_images(knn_model.query(bird))['image'].show()
```

# Using lambda to find and show nearest neighbor images

```python
near_neighbors = lambda i: get_images(knn_model.query(image_train[i:i+1]))['image'].show()
```
